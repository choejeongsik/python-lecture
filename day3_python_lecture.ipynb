Example1

두개의 박스가 있다. 작은 박스는 5개가 들어가고 큰 박스는 10개가 들어간다. 큰박스의 무게는 8kg이고 작은 박스의 무게는 5kg이다.
입력한 수를 포장할때 최저 무게가 되는 큰박스의 개수, 작은박스의 개수, 총 무게를 출력하라

```python
x=int(input())

big=0
small=0

big=x//10

remain=x-big*10

if (remain>=6):
    big=big+1
else:
    small=1

print(big)
print(small)
print(8*big+5*small+x)
```

Example2

동전의 개수를 최소로 만드는 동전교환기를 만들어라

```python
x=int(input())

a=x//500
b=(x-500*a)//100
c=(x-500*a-100*b)//50
d=(x-500*a-100*b-50*c)//10
e=x-500*a-100*b-50*c-10*d

print(a, b, c, d, e)
```

Example3

두개의 층을 입력받고 모두 점등된 호라인의 개수와 하나라도 점등된 호라인의 개수를 출력하라
789564.png

```python
floorlist={1:4, 2:8, 3:12, 4:16, 5:20, 6:24, 7:28, 8:32, 9:36}

x=int(input())
y=int(input())

a=floorlist[x]
b=floorlist[y]

ans1=bin(a&b)
ans2=bin(a|b)

print(ans1.count('1'))
print(ans2.count('1'))
```
