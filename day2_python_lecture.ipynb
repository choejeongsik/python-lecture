Example6
문자열에서 어떤 문자가 나올 확률을 구하시오, 문자가 나올 확률의 평균으로 계산된다.

```python
dic={'a':7, 'b':9, 'c':11, 'd':13, 'e':3, 'f':20, 'g':5, 'h':30,'i':2, 'j':5}
x=input()
xlist=list(x)
temp=0

for element in xlist:
    ans=temp+dic[element]
    temp=ans
print(ans/len(xlist))
```

Example7 3명의 학생의 국어 영어 수학 점수를 저장하고 학생 번호(1, 2, 3)를 입력 받으면 점수를 올림차순으로 정렬하고 최고점과 최하점을 출력하라

```python
a1=int(input())
b1=int(input())
c1=int(input())
list1=[a1, b1, c1]

a2=int(input())
b2=int(input())
c2=int(input())
list2=[a2, b2, c2]

a3=int(input())
b3=int(input())
c3=int(input())
list3=[a3, b3, c3]

total={1:list1, 2:list2, 3:list3}

g=int(input())
result=total[g]
result.sort()
print(result, max(result), min(result))
```
